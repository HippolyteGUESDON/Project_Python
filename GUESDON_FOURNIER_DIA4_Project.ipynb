```python
#Convert categorical variable into dummy/indicator variables
data_dummies = pd.get_dummies(df, prefix=['S','Family','High_caloric','Snack','Smoke','Count_calories','Alcool','Traveling'], columns=['Gender','Family_history','High_caloric_food','Snacking','Smoke','Count_calories','Alcool','Traveling_way'])
#display(data_dummies.columns)

# convert class Shape in numerical data
data_dummies["Shape"].replace({'Insufficient_Weight':0,'Normal_Weight' : 1,'Overweight_Level_I' : 2,'Overweight_Level_II' : 3,'Obesity_Type_I':4,'Obesity_Type_II':5,'Obesity_Type_III':6},inplace=True)
```

```python
#Separate to have the features and the prediction
X = data_dummies.iloc[:, data_dummies.columns != 'Shape'].values
y = data_dummies.iloc[:, 8].values
```

```python
from sklearn.model_selection import train_test_split
#Doing the split with test and train data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
#Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#Classifier implementing the k-nearest neighbors vote
classifier = KNeighborsClassifier(n_neighbors=5)
#Learn with the train data
classifier.fit(X_train, y_train)
```

```python
#classifier is applied to the X_test and the prediction are enter in y_pred
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
#We have the confusion matrix
print(confusion_matrix(y_test, y_pred))
#We have the classification report with the accuracy
print(classification_report(y_test, y_pred))
```
